```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import Birch
import pandas as pd
```

```python
df = pd.read_csv('aggregation.csv')
```

```python
x = df.iloc[:,[0,1]].values
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
brc = Birch(branching_factor=20, n_clusters=7, threshold=1.5)
brc.fit(x)
```

```python
labels = brc.predict(x)
labels
```

```python
plt.scatter(x[:,0], x[:,1], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
```
