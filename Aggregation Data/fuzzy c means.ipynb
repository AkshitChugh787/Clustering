```python
from fcmeans import FCM
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('aggregation.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,[0,1]].values
```

```python
x
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x[:,0],x[:,1],color = 'b')
plt.show()
```

```python
fcm = FCM(n_clusters=7)
fcm.fit(x)
```

```python
fcm_centers = fcm.centers
fcm_labels = fcm.predict(x)
```

```python
fcm_centers
```

```python
fcm_labels
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x[:,0],x[:,1],c=fcm_labels)
plt.show()
```
