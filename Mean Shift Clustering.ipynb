```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift, estimate_bandwidth
```

```python
x, y = make_blobs(n_samples=2000, n_features = 4, centers = [(5,5), (3,3), (1,1),(3,6),(6,1)], cluster_std = 0.50)
```

```python
plt.scatter(x[:, 0], x[:, 1])
plt.xlabel("X - axis ")
plt.ylabel("Y - axis ")
plt.show()
```

```python
z = estimate_bandwidth(x, quantile=0.2, n_samples=500)
```

```python
model = MeanShift(bandwidth=z)
model.fit(x)
y_pred  = model.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap="viridis")
plt.xlabel("X - axis ")
plt.ylabel("Y - axis ")
```
